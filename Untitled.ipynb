```python
# importar librería
from prophet import Prophet
```

```python
# Función encargada de hacer predicciones de una tienda y de un periodo específico.

def sales_predictions(store_id, pred_sales_df, periods):
    
    pred_sales_df = pred_sales_df[pred_sales_df.Store == store_id]
    pred_sales_df = pred_sales_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})
    
    pred_sales_df = pred_sales_df.sort_values('ds')
    
    
    # Modelado
    model = Prophet()
    model.fit(pred_sales_df)
    
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)
    
    
    # Gráfico
    figure = model.plot(forecast, xlabel='Fechas', ylabel='Ventas')
    figure2 = model.plot_components(forecast)
```

```python
df = sales_predictions(10, store_sales_df, 60)
```

```python
# Modelado con vacaciones

def sales_predictions(store_id, pred_sales_df, holidays, periods):
    
    pred_sales_df = pred_sales_df[pred_sales_df.Store == store_id]
    pred_sales_df = pred_sales_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})
    
    pred_sales_df = pred_sales_df.sort_values('ds')
    
    
    # Modelado
    model = Prophet(holidays=holidays)
    model.fit(pred_sales_df)
    
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)
    
    
    # Gráfico
    figure = model.plot(forecast, xlabel='Fechas', ylabel='Ventas')
    figure2 = model.plot_components(forecast)
```

```python
# dataset con vacaciones de escuela (Solo las fechas de cuando hay vacaciones escolares)
school_holidays = store_sales_df[store_sales_df.SchoolHoliday == 1].loc[:,'Date'].values

school_holidays = np.unique(school_holidays)
```

```python
# Dataset con dias festivos
state_holidays = store_sales_df[store_sales_df.StateHoliday != "0"].loc[:,'Date'].values

state_holidays = np.unique(state_holidays)
```

```python
# Dataframe vacaciones de escuela para Prophet
school_holidays = pd.DataFrame({'ds':pd.to_datetime(school_holidays), 'holiday': 'school_holiday'})
```

```python
school_holidays.head()
```

```python
# Dataframe dia festivos para Prophet
state_holidays = pd.DataFrame({'ds':pd.to_datetime(state_holidays), 'holiday': 'state_holidays'})
```

```python
state_holidays.head()
```

```python
holidays_df = pd.concat((state_holidays, school_holidays), axis=0)
```

```python
holidays_df
```

```python
df = sales_predictions(6, store_sales_df, holidays_df, 90)
```
